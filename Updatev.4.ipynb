```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

# โหลดไฟล์ CSV
file_path = "/content/Updated_Tourist_OneHot_YearRound_Corrected.csv"
df = pd.read_csv(file_path)

# ตรวจสอบคอลัมน์ที่ไม่ใช่ตัวเลข
non_numeric_cols = ['Destination', 'Country', 'Famous Attraction', 'Attractive Activities', 'Activity Category']

# แปลงข้อความเป็นตัวเลขด้วย Label Encoding
label_encoders = {}
for col in non_numeric_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # เก็บ Encoder ไว้เผื่อใช้ภายหลัง

# ---- ทำ One-Hot Encoding สำหรับเดือน ----
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
df = pd.get_dummies(df, columns=months)  # ทำ One-Hot Encoding

# กำหนด Features และ Target
X = df.drop(columns=['Activity Category'])  # ฟีเจอร์ (ลบ Target ออก)
y = df['Activity Category']  # Target (สิ่งที่ต้องการพยากรณ์)

# ปรับค่าให้เหมาะสมกับการเทรนโดยใช้ StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# แบ่งข้อมูลเป็น Train และ Test (80% เทรน, 20% ทดสอบ)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ---- สร้างโมเดล FNN ----
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # Hidden Layer 1
    keras.layers.Dense(32, activation='relu'),  # Hidden Layer 2
    keras.layers.Dense(len(y.unique()), activation='softmax')  # Output Layer (ใช้ Softmax)
])

# คอมไพล์โมเดล
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# เทรนโมเดล
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=30, batch_size=16, verbose=1)

# ---- พล็อตกราฟ Accuracy และ Loss ----
plt.figure(figsize=(12, 5))

# กราฟ Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy')

# กราฟ Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')

plt.show()

# ---- ประเมินผลลัพธ์ ----
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"\nFinal Test Accuracy: {test_acc:.4f}")
print(f"Final Test Loss: {test_loss:.4f}")

# ---- สร้าง Heatmap 1: ความนิยมของกิจกรรมในแต่ละประเทศ ----
heatmap1_data = df.pivot_table(index="Country", columns="Attractive Activities", aggfunc="size", fill_value=0)

plt.figure(figsize=(12, 6))
plt.imshow(heatmap1_data, cmap="YlGnBu", aspect='auto')
plt.colorbar(label="Frequency")
plt.xticks(ticks=np.arange(len(heatmap1_data.columns)), labels=heatmap1_data.columns, rotation=90)
plt.yticks(ticks=np.arange(len(heatmap1_data.index)), labels=heatmap1_data.index)
plt.title("Heatmap: นักท่องเที่ยวนิยมทำกิจกรรมอะไรในแต่ละประเทศ")
plt.xlabel("กิจกรรม")
plt.ylabel("ประเทศ")
plt.show()

# ---- สร้าง Heatmap 2: ความนิยมของประเทศในแต่ละเดือน ----
# ตรวจสอบคอลัมน์ที่เกี่ยวข้องกับเดือน
month_cols = [col for col in df.columns if any(month in col for month in months)]
print("Columns detected as months:", month_cols)  # ตรวจสอบชื่อคอลัมน์เดือน

# รวมข้อมูลในแต่ละประเทศตามเดือน
heatmap2_data = df.groupby("Country")[month_cols].sum()

# แสดงข้อมูลที่รวมกัน
print(heatmap2_data.head())  # ตรวจสอบค่าที่รวมกัน

# สร้าง Heatmap สำหรับความนิยมของประเทศในแต่ละเดือน
plt.figure(figsize=(12, 6))
plt.imshow(heatmap2_data, cmap="Reds", aspect='auto')
plt.colorbar(label="Popularity")
plt.xticks(ticks=np.arange(len(months)), labels=months, rotation=90)
plt.yticks(ticks=np.arange(len(heatmap2_data.index)), labels=heatmap2_data.index)
plt.title("Heatmap: ความนิยมของประเทศในแต่ละเดือน")
plt.xlabel("เดือน")
plt.ylabel("ประเทศ")
plt.show()

print(df.columns)
print(df.head())
```
